````python
import pandas as pd 
import numpy as np
````

````python
df = pd.DataFrame(np.random.randn(11, 9),
                   index=list('abcdef12345'),
                   columns=list('ABCD12345'))
````

````python
df
````

# 	**1.label-based indexing(.loc)**
   
```
df.loc[row_indexer,column_indexer]
```


> 1.1.	A single label

> 1.2.	A list or array of labels

> 1.3.	A slice object with labels 

> 1.4.	A boolean array 


> 1.5.	A callable function with one argument 



````python
# a single label
df1=df.loc['a', '5']
# a list / array of labels
df2=df.loc[['a', 'b', 'c'], ]
# a slice object with labels
df3=df.loc['a':'b',]
# a boalean array
df4=df.loc['a']>5
# a callabel function 
df5=df.loc[lambda df: df['A'] > 0, :]
````

# 	**2.	Integer/position-based indexing(.iloc)**


```
df.loc[row_indexer,column_indexer]
```




> 2.1.	A single label

> 2.2.	A list or array of labels

> 2.3.	A slice object with labels 

> 2.4.	A boolean array 


> 2.5.	A callable function with one argument 



````python
# a single label
df6=df.iloc[5]
# a list / array of labels
df7=df.iloc[[3, 4, 5],]
# a slice object with labels
df8=df.iloc[7:1,]
# a boalean array
df9=df.iloc[5]>5
# a callabel function 
df10=df.iloc[:, lambda df: [0, 1]]
````
